<h4>Introduction/Business Problem</h4>

I was interested in the idea of running a cluster analysis on restaurant location data in Los Angeles to see if I could discern patterns in what types of restaurants and other eateries appear in certain neighborhoods as well as the volume in which they appear. Los Angeles is one of the premier food scenes in the United States, and its restaurant ecosystem is incredibly competetitive. Location is extremely important to the success of a restaurant, and even a marginal advantage gleaned from this sort of cluster analysis could be critical to the survival of a business. These clusters could reveal certain areas to avoid due to oversaturation of a particular food group or deserted areas that could present a market opportunity for certain styles of restaurants. A restaurant trade association could, perhaps, provide this data analysis to new businesses to help them succeed in the Los Angeles market and limit inefficient, unwanted restaurant turnover.
